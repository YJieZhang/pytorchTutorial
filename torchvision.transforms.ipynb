```python
imgPath = './data/chair.jpg'
img = Image.open(imgPath).convert('RGB')
plt.imshow(img)
plt.show()
```

```python
# Resize
# Example: (800,800)-->Resize((300,300))-->(300,300)
#          (800,500)-->Resize(250)-->(400,250)
# Default: PIL.Image.BILINEAR interpolation
transform = transforms.Resize((5000,5000))
plt.imshow(transform(img))
plt.show()
```

```python
# Center Crop
# params: (height, width), or square_size
transform = transforms.CenterCrop((300,300))
plt.imshow(transform(img))
plt.show()
```

```python
# Random Crop
# params: similar to Center Crop
#         padding to the border
transform = transforms.RandomCrop((700,700), padding=100)
plt.imshow(transform(img))
plt.show()
```

```python
# Random Horizontal Flip
transform = transforms.RandomHorizontalFlip()
plt.imshow(transform(img))
plt.show()
```

```python
# Random Vertical Flip
transform = transforms.RandomVerticalFlip()
plt.imshow(transform(img))
plt.show()
```

```python
# Random Resized Crop
transform = transforms.RandomResizedCrop(800)
plt.imshow(transform(img))
plt.show()
```

```python
# Grayscale
# Default: num_out_channels=1
transform = transforms.Grayscale(num_output_channels=3)
plt.imshow(transform(img))
plt.show()
```

```python
# Random Grayscale
# num_out_channels is same as num_in_channels
transform = transforms.RandomGrayscale(p=0.1)
plt.imshow(transform(img))
plt.show()
```

```python
# Pad
# Default: fill=0
transform = transforms.Pad(50,fill=(0,255,0))
plt.imshow(transform(img))
plt.show()
```

```python
# Color Jitter
# birghtness, contrast, saturation, hue
transform = transforms.ColorJitter(hue=0.25)
plt.imshow(transform(img))
plt.show()
```

```python
# Random Rotation
transform = transforms.RandomRotation(45, resample=Image.BILINEAR, expand=False, center=None)
plt.imshow(transform(img))
plt.show()
```
